# Bayes

$\Phi(\mu,\sigma)是正态分布密度函数$  
$p(x|\omega_1) = 0.2 * \Phi(0,1) + 0.4 * \Phi(4,0.8) + 0.4 * \Phi(7,3)$  
$p(x|\omega_2) = 0.2 * \Phi(3,2) + 0.8 * \Phi(9,3.5)$   
$P(\omega_1) = 0.35$  

## 1.求$P(e)$的最小分界面

```python
import numpy as np
from scipy.stats import norm
import numpy.random as rnd
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.colors as mcolors

def make_colormap(seq):
    """Return a LinearSegmentedColormap
    seq: a sequence of floats and RGB-tuples. The floats should be increasing
    and in the interval (0,1).
    """
    seq = [(None,) * 3, 0.0] + list(seq) + [1.0, (None,) * 3]
    cdict = {'red': [], 'green': [], 'blue': []}
    for i, item in enumerate(seq):
        if isinstance(item, float):
            r1, g1, b1 = seq[i - 1]
            r2, g2, b2 = seq[i + 1]
            cdict['red'].append([item, r1, r2])
            cdict['green'].append([item, g1, g2])
            cdict['blue'].append([item, b1, b2])
    return mcolors.LinearSegmentedColormap('CustomMap', cdict)
```

```python
p_w1 = 0.35
p_w2 = 1 - p_w1

x = np.linspace(-10, 25, 500)
y1 = 0.2 * norm.pdf(x, 0, 1) + \
     0.4 * norm.pdf(x, 4, 0.8) + \
     0.4 * norm.pdf(x, 7, 3)
y2 = 0.2 * norm.pdf(x, 3, 2) + \
     0.8 * norm.pdf(x, 9, 3.5)
z1 = p_w1 * y1
z2 = p_w2 * y2
```

后验概率$p(w_1|x), p(w_2|x)$

```python
plt.figure(figsize=(7,7))
plt.plot(x,z1,label=r'$p(w_1|x)$')
plt.plot(x,z2,label=r'$p(w_2|x)$')
plt.legend()
```

```python
z_delta = z1 - z2
z_delta.shape
flag = []   
for i in range(z_delta.shape[0] - 1):
    if z_delta[i] * z_delta[i+1] < 0:
        flag.append(i);
```

```python
y = np.linspace(0, 0.1, 500)
xv, yv = np.meshgrid(x, y)
X = np.c_[xv.ravel(), yv.ravel()]
# print(xv.shape,yv.shape,X.shape)
z = np.zeros(X.shape[0])
for i in range(X.shape[0]):
    #for j in range(100):
    tx,ty = X[i]
    if (ty<z1[i%500] or ty<z2[i%500]):
        z[i] = 1
    if (ty<z1[i%500] and ty<z2[i%500]) and (z1[i%500] < z2[i%500]):
        z[i] = 2
    if (ty<z1[i%500] and ty<z2[i%500]) and (z2[i%500] < z1[i%500]):
        z[i] = 3      
z = z.reshape(xv.shape)

c = mcolors.ColorConverter().to_rgb
mc = make_colormap([c('white'), 0.25, c('yellow'), 0.5, c('deepskyblue'),0.75,c('lawngreen')])

plt.figure(figsize=(7,7))
plt.plot(x,z1,label=r'$p(x|w_1)p(w_1)$')
plt.plot(x,z2,label=r'$p(x|w_2)p(w_2)$')
plt.contourf(xv,yv,z,cmap=mc)
plt.legend()
print("分界面\n横坐标:", x[flag])
```

由上图和计算结果可知分类面分别为$x=-2.91583166, 0.94188377, 2.9759519, 4.93987976$  
其中$x \in [-10,-2.9], x \in [0.94, 2.97], x \in [4.93, 25]$ 属于 $w_1$类, 剩余的区间属于$w_2$类

## 2. 求$P(e)=0.2$时,使得$P_2(e)$最小的决策面

```python
l = y1/y2
pe_1 = 0.2
plt.figure(figsize=(7,7))
plt.plot(x,l,label="likelihood")
dl = np.linspace(0,l.max(),500)
for t in range(500):
    e_y1 = y1[l < dl[t]]
    e_y2 = y2[l > dl[t]]
    if (e_y1.sum()*0.07 > 0.195 and e_y1.sum()*0.07 < 0.205):
        gl = l < (dl[t] + 0.01)
        sl = l > (dl[t] - 0.006)
        break
    
hyperplane = [x[gl & sl][0],x[gl & sl][-1]]

plt.plot(np.ones(500) * hyperplane[0],np.linspace(0,8,500),color="#000000",label="hyperplane1")
plt.plot(np.ones(500) * hyperplane[1],np.linspace(0,8,500),color="#5e5e5e",label="hyperplane2")
plt.plot(np.linspace(-10, 25, 500), np.ones(500) * dl[t], label="$\lambda$")
plt.legend()
```

```python
print(dl[t],e_y1.sum()*0.07,e_y2.sum()*0.07)
print(hyperplane)
```

由上图和结果可知$\lambda$的取值大约为$0.6493$,所得到的决策面为$x = -3.68, 6.97 \quad p_1(e) \approx 0.20 \quad p_2(e) \approx 0.416$  
下图绘画出最后的分类结果

```python
plt.figure(figsize=(7,7))

x1 = x<hyperplane[0]
x2 = (x > hyperplane[0]) & (x < hyperplane[1])
x3 = x>hyperplane[1]

fill_x = np.append(x[x1],np.append(x[x2],x[x3]))
fill_z1 = np.append(z1[x1],np.append(np.zeros(x[x2].shape[0]),z1[x3]))
fill_z2 = np.append(np.zeros(x[x1].shape[0]),np.append(z2[x2],np.zeros(x[x3].shape[0])))

plt.plot(x,z1,label=r'$p(x|w_1)p(w_1)$')
plt.plot(x,z2,label=r'$p(x|w_2)p(w_2)$')
plt.fill(fill_x, fill_z1,fill_x, fill_z2)
plt.legend()
```

## 3. 最小风险贝叶斯决策
如果风险损失表是:

|  $\quad$  | $w_1$ | $w_2$ |
|   :---:   | :---: | :---: |
|$\alpha_1$ |   0   |   2   |
|$\alpha_2$ |   1   |   0   |

求最小风险损失下的分界面:

```python
z1 = 2 * p_w2 * y2 / (p_w1 * y1 + p_w2 * y2)
z2 = p_w1 * y1 / (p_w1 * y1 + p_w2 * y2)
plt.figure(figsize=(7,7))
plt.plot(x,z1,label=r'$p(w_1|x)$')
plt.plot(x,z2,label=r'$p(w_2|x)$')


z_delta = z1 - z2
z_delta.shape
flag = []   
for i in range(z_delta.shape[0] - 1):
    if z_delta[i] * z_delta[i+1] < 0:
        flag.append(i);
print(x[flag])

plt.plot(np.ones(500) * x[flag][0],np.linspace(0,2,500),color="#000000",label="hyperplane1")
plt.plot(np.ones(500) * x[flag][1],np.linspace(0,2,500),color="#5e5e5e",label="hyperplane2")
plt.legend()
```

由上图和计算结果可知分类面分别为$x=-2.28456914,0.31062124$  
其中$x \in [-2.28, 0.31]$ 属于 $w_1$类, 剩余的区间属于$w_2$类
